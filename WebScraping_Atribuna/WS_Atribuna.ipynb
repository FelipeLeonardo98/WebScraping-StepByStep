# **Case Example: WebScraping - A Tribuna News**

### **Importing Libs and dependencies**

```python
# WebScraping libs
import requests
from bs4 import BeautifulSoup

# Pandas
import pandas as pd

# Date
from datetime import date
today = date.today()

# Numpy
import numpy as np
```

### **Setting dataFrame**

```python
# Setting dataFrame
df_news = pd.DataFrame(columns=['Title','Subheading','Date'])

# Insertfunction
def insert(df, row):
    insert_loc = df_news .index.max()

    if np.isnan(insert_loc):
        df_news.loc[0] = row
    else:
        df_news .loc[insert_loc + 1] = row


display(df_news)
```

### **WebScraping**

```python
# initials variables
response = requests.get('https://www.atribuna.com.br')
content = response.content
site = BeautifulSoup(content, 'html.parser')

# all structure (father element)
raws = site.findAll('div', attrs={'class': 'pack3'})

for new in raws:
  title = new.find('h1', attrs={'class':'titulo'})
  subheading = new.find('h1',attrs={'class':'rel'})
  
  # checking
  if(title and subheading):
    print(title.text)
    print(subheading.text)
    # insert into df
    insert(df_news ,[title.text,subheading.text,today])
    print('------------------------------------------------------------------------------------')
    print()
  elif(title):
    print(title.text)

    # insert into df
    insert(df_news ,[title.text,'Do not have a subheading',today])
    print('------------------------------------------------------------------------------------')
    print()

# check
display(df_news)
url = 'C:/Users/ferra/Documents/Python Scripts/WebScraping_Atribuna'
df_news.to_excel(f'{url}/news_{today}.xlsx', sheet_name='news')
```
